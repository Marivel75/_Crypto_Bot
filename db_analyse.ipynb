```python
# Notebook d'analyse de la base de données Crypto Bot
# Visualisation des données OHLCV sous forme de DataFrames pandas

import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
#plt.style.use('seaborn')
pd.set_option('display.max_columns', None)

# Ajouter le chemin racine
sys.path.append(os.path.dirname(os.path.abspath('')))

from sqlalchemy import create_engine, inspect
from src.config import settings

print('Configuration terminée')
```

# Analyse de la Base de Données Crypto Bot

Visualisation des données OHLCV depuis PostgreSQL.

```python
# Connexion à la base de données
engine = create_engine(f'postgresql+psycopg2://{settings.POSTGRES_USER}:{settings.POSTGRES_PASSWORD}@{settings.POSTGRES_HOST}:{settings.POSTGRES_PORT}/{settings.POSTGRES_DB}')
inspector = inspect(engine)
print('Connexion établie')
print('-'*45)

# Vérifier la structure
print('Analyse de la structure de la table ohlcv:')
if inspector.has_table('ohlcv'):
    columns = inspector.get_columns('ohlcv')
    print(f'Table ohlcv existe avec {len(columns)} colonnes')
    
    # Afficher la structure
    structure_df = pd.DataFrame([{"Colonne": col['name'], "Type": str(col['type']), "Nullable": col['nullable']} for col in columns])
    display(structure_df)
else:
    print('Table ohlcv non trouvée')
print('-'*45)

# Charger les données
print('Chargement des données OHLCV...')
df = pd.read_sql('SELECT * FROM ohlcv ORDER BY timestamp DESC LIMIT 1000', engine)
print(f'Données chargées: {len(df)} enregistrements')
display(df.head())
print('-'*45)

# INfo géné
print('Informations sur le DataFrame:')
df.info()
# Statistiques
print('\n')
print('Statistiques descriptives:')
display(df.describe())



# Analyse par symbole
print('Analyse par sympbole:')
symbol_stats = df.groupby('symbol').agg({"timestamp": "count", "open": "mean", "volume": "sum"})
symbol_stats.columns = ['count', 'avg_price', 'total_volume']
display(symbol_stats.sort_values('count', ascending=False))
print('-'*45)

# Visualisation des prix
print('Visualisation des prix de clôture pour les 3 premiers symboles:')
plt.figure(figsize=(12, 6))
for symbol in df['symbol'].unique()[:3]:
    symbol_data = df[df['symbol'] == symbol].sort_values('timestamp')
    plt.plot(symbol_data['timestamp'], symbol_data['close'], label=symbol)
plt.title('Évolution des prix')
plt.legend()
plt.show()


# Fermer la connexion
print('Fermeture de la connexion à la base de données...')
engine.dispose()
print('Analyse terminée')
```
